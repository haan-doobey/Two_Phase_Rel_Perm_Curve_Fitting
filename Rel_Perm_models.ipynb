# 2 Phase Relative Permeability Models



1. **Corey's 2 Phase Model - Valid for Drainage in a Consolidated Rock.**

$Swn = \frac {Sw - Swirr}{1 - Swirr}  $

$Krw = Swn^4$

$Kro = (1 - Swn)^2 * (1 - Swn^2)$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Sw = np.linspace(0,1,100)

Swirr = 0.20

Swn = (Sw - Swirr)/(1-Swirr)

Krw = Swn**4

Kro = ((1 - Swn)**2)*(1 - Swn**2)

plt.style.use('default')

# Plot water rel perm
plt.plot(Sw, Krw, label='Krel-water (Wetting P.)', linewidth=3)

# Plot oil rel perm
plt.plot(Sw, Kro, label='Krel-oil (Non-Wetting P.)', linewidth=3)

plt.title("Rel-Perm by Corey's Model")
plt.xlabel('Sw')
plt.ylabel('Krel(o,w)')
plt.grid()
plt.legend(loc='best')
```

2. **Naar-Henderson's Model**

$Swn = \frac {Sw - Swirr}{1 - Swirr}  $

$Krw = Swn^4$

$Kro = (1 - 2*Swn)^(\frac{3}{2}) * [ 2 - \sqrt (1 - 2*Swn)]$

```python
Swn = (Sw - Swirr)/(1-Swirr)

Krw = Swn**4

Kro_h = ((1 - 2*Swn)**(1.5))*(2 - (1-2*Swn)**(0.5))

plt.style.use('default')
plt.plot(Sw,Krw,label='Krel-water(Wetting P.)',linewidth=3,color='green')
plt.plot(Sw,Kro_h,color='r',label='Krel-Oil(Non-Wetting P.)',linewidth=3)
plt.title("Rel-Perm by Naar-Henderson's Model")
plt.xlabel('Sw')
plt.ylabel('Krel(o,w)')
plt.grid()
plt.legend(loc='best')
```
